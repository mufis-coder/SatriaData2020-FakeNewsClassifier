```python
import nltk
import re
nltk.download('stopwords')
nltk.download('punkt')
from nltk.corpus import stopwords
from nltk import sent_tokenize, word_tokenize
```

```python
from nltk.corpus import stopwords
# ----------------------- get stopword from NLTK stopword -------------------------------
# get stopword indonesia
list_stopwords = stopwords.words('indonesian')
stop_words = list_stopwords
```

```python
def HiSed(df, perintah, stop_words):
    if perintah == 'sederhana':
        df_return = pd.DataFrame()
    elif perintah == 'hitung':
        dfCounts_return = dict()
        
    for index, row in df.iterrows():
        filter_sentence = ''
        sentence = row['total']
        # Cleaning dengan regular expression
        sentence = re.sub(r'[^\w\s]', '', sentence)
        # Tokenization
        words = nltk.word_tokenize(sentence)
        # Menghilangkan stopwords
        words = [w for w in words if not w in stop_words]
        if perintah == 'sederhana':
            # Lower case
            for words in words:
                filter_sentence = filter_sentence  + ' ' + words.lower()
            df_return.loc[index, 'total'] = filter_sentence
        elif perintah == 'hitung':
            for word in words:
                if word.lower() in dfCounts_return:
                    dfCounts_return[word.lower()] += 1
                else:
                    dfCounts_return[word.lower()] = 1
            
    if perintah == 'sederhana':
        return df_return
    elif perintah == 'hitung':
        return dfCounts_return
```

```python
dict_kata_train = HiSed(df, 'hitung', stop_words)
```

```python
df_kata_train = pd.DataFrame({'kata':[x for x, y in dict_kata_train.items()], 
                              "jumlah":[y for x, y in dict_kata_train.items()]})
df_kata_train.sort_values(by=['jumlah'], inplace=True, ascending=False)
df_kata_train = df_kata_train.reset_index(drop=True)
```

```python
df_kata_train.head(10)
```

```python
dict_kata_test = HiSed(df_t, 'hitung', stop_words)
```

```python
df_kata_test = pd.DataFrame({'kata':[x for x, y in dict_kata_test.items() if x not in dict_kata_train], 
                             "jumlah":[y for x, y in dict_kata_test.items() if x not in dict_kata_train]})
df_kata_test.sort_values(by=['jumlah'], inplace=True, ascending=False)
df_kata_test = df_kata_test.reset_index(drop=True)
```
