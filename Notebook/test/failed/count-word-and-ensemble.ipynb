```python
df_counts = dict()
#preprocessing data training
for index, row in df.iterrows():
    filter_sentence = ''
    sentence = row['total']
    # Cleaning dengan regular expression
    sentence = re.sub(r'[^\w\s]', '', sentence)
    # Tokenization
    words = nltk.word_tokenize(sentence)
    # Menghilangkan stopwords
    words = [w for w in words if not w in stop_words]
    # Lower case
    for word in words:
        if word.lower() in df_counts:
            df_counts[word.lower()] += 1
        else:
            df_counts[word.lower()] = 1
        filter_sentence = filter_sentence  + ' ' + word.lower()
    
    df.loc[index, 'total'] = filter_sentence
    
df_t_counts = dict()
#preprocessing data testing    
for index, row in df_t.iterrows():
    filter_sentence = ''
    sentence = row['total']
    # Cleaning dengan regular expression
    sentence = re.sub(r'[^\w\s]', '', sentence)
    # Tokenization
    words = nltk.word_tokenize(sentence)
    # Menghilangkan stopwords
    words = [w for w in words if not w in stop_words]
    # Lower case
    for word in words:
        if word.lower() in df_t_counts:
            df_t_counts[word.lower()] += 1
        else:
            df_t_counts[word.lower()] = 1
        filter_sentence = filter_sentence  + ' ' + word.lower()
    
    df_t.loc[index, 'total'] = filter_sentence
```

```python
df_counts_df = pd.DataFrame({'kata':[x for x, y in df_counts.items()], "jumlah":[y for x, y in df_counts.items()]})
df_counts_df.sort_values(by=['jumlah'], inplace=True, ascending=False)
df_counts_df = df_counts_df.reset_index(drop=True)
```

```python
df_t_counts_df = pd.DataFrame({'kata':[x for x, y in df_t_counts.items()], "jumlah":[y for x, y in df_t_counts.items()]})
df_t_counts_df.sort_values(by=['jumlah'], inplace=True, ascending=False)
df_t_counts_df = df_t_counts_df.reset_index(drop=True)
```

```python
display(df_t_counts_df)
```

```python
kata_train_useless = df_counts_df[df_counts_df['jumlah']<=2]
kata_test_useless = df_t_counts_df[df_t_counts_df['jumlah']<=2]
```

```python
kata_train_useless = [y['kata'] for x, y in kata_train_useless.iterrows()]
kata_test_useless = [y['kata'] for x, y in kata_test_useless.iterrows()]
```

```python
#preprocessing data training
for index, row in df.iterrows():
    filter_sentence = ''
    sentence = row['total']
    # Cleaning dengan regular expression
    sentence = re.sub(r'[^\w\s]', '', sentence)
    # Tokenization
    words = nltk.word_tokenize(sentence)
    # Menghilangkan stopwords
    words = [w for w in words if not w in stop_words]
    # Lower case
    for word in words:
        if word.lower() not in kata_train_useless:
            filter_sentence = filter_sentence  + ' ' + word.lower()
    df.loc[index, 'total'] = filter_sentence
    
#preprocessing data testing    
for index, row in df_t.iterrows():
    filter_sentence = ''
    sentence = row['total']
    # Cleaning dengan regular expression
    sentence = re.sub(r'[^\w\s]', '', sentence)
    # Tokenization
    words = nltk.word_tokenize(sentence)
    # Menghilangkan stopwords
    words = [w for w in words if not w in stop_words]
    # Lower case
    for word in words:
        if word.lower() not in kata_test_useless:
            filter_sentence = filter_sentence  + ' ' + word.lower()
    df_t.loc[index, 'total'] = filter_sentence
```

```python
#inisiasi x dan y
x_train = df['total']
y_train = df['label']
```

```python
x_test = df_t['total']
y_test = df_t['label']
```

```python
# Initialize a TfidfVectorizer
tfidf_vectorizer=TfidfVectorizer(max_df=1.0)

#Fit and transform train set, transform test set
tfidf_train=tfidf_vectorizer.fit_transform(x_train)
tfidf_test=tfidf_vectorizer.transform(x_test)
```

```python
tfidf_train_array = tfidf_train.toarray()
tfidf_test_array = tfidf_test.toarray()
tfidf_train_df = pd.DataFrame(tfidf_train_array)
tfidf_test_df = pd.DataFrame(tfidf_test_array)
display(tfidf_train_df.head())
display(tfidf_test_df.head())
```

```python
subset4_x, subset4_y = tfidf_train_df, y_train
```

```python
from sklearn.ensemble import RandomForestClassifier

# Create the model with 100 trees
model = RandomForestClassifier(n_estimators=100, 
                               bootstrap = True,
                               max_features = 'sqrt')
# Fit on training data
model.fit(subset4_x, subset4_y)
y_pred4 = model.predict(tfidf_test_df)
score4 = metrics.f1_score(y_test, y_pred4)
print(f'F1 Score: {round(score4*100,2)}%')
```
